# Conclusión final

El EarlyStopping no mejoró el resultado, al contrario. Cortó en el epoch 4 con un accuracy en test de 0.8328, peor que el LSTM sin EarlyStopping (0.8512) e incluso peor que la regresión logística (0.8728).
Mirando las curvas, el val_loss tocó su mejor punto en el epoch 2 (0.3266) y luego empezó a subir. El patience=2 hizo que el entrenamiento siguiera dos epochs de más antes de parar, y aunque restore_best_weights debería recuperar los pesos de ese mejor punto, el resultado en test salió peor de lo esperado. Posiblemente con un patience más bajo (1) o monitorizando val_accuracy en vez de val_loss el resultado habría sido distinto.
En cualquier caso, el ranking final no cambia: la regresión logística con TF-IDF sigue siendo la mejor opción para este dataset, y queda claro que para 5.000 muestras un modelo simple le saca ventaja a una red neuronal, incluso después de intentar corregirla.
